```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
books = pd.read_csv('Books.csv')
users = pd.read_csv('Users.csv')
rating  = pd.read_csv('Ratings.csv')
```

```python
print(books.columns)
print(books.shape)
books.head(2)
```

```python
del books['Image-URL-S']
del books['Image-URL-L']
```

```python
books.rename(columns={'Book-Title': 'title' , 'Book-Author':'author' , 'Year-Of-Publication':'year' ,'Publisher' :'publisher','Image-URL-M':'url'} , inplace = True)
```

```python
books.head(2)
```

```python
rating.head(2)
```

```python
rating.rename(columns={'User-ID':'user' , 'Book-Rating' : 'rating'},inplace = True)
rating.head(2)
```

```python
rating_name = rating.merge(books , on='ISBN')
rating_name.head(2)
```

```python
x = rating_name.groupby('user').count()['rating'] > 200
padhe_likhe_users = x[x].index
```

```python
rating_name = rating_name[rating_name['user'].isin(padhe_likhe_users)]
```

```python
rating_name.shape
```

```python
y = rating_name.groupby('title').count()['rating']>=50
famous_books = y[y].index
```

```python
final_rating = rating_name[rating_name['title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='title',columns='user',values='rating')
pt.fillna(0,inplace=True)
pt
```

```python
np.where(pt.index == 'Harry Potter and the Chamber of Secrets (Book 2)')[0][0]
```

```python
np.where(pt.index == '1984')
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(pt)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distances , suggestions = model.kneighbors(pt.iloc[230].values.reshape(1,-1), n_neighbors=6)
```

```python
for i in range(len(suggestions[0])):
  print(pt.index[suggestions[0][i]] , end='\n')
```

```python
def searchBook(book):
  bookId = np.where(pt.index == book)[0][0]
  distances , suggestions = model.kneighbors(pt.iloc[bookId].values.reshape(1,-1), n_neighbors=6)
  books = []
  for i in range(len(suggestions[0])):
    if i != 0:
      # print(pt.index[suggestions[0][i]])
      books.append(pt.index[suggestions[0][i]])
  return books
```

```python
recbooks = searchBook('Harry Potter and the Prisoner of Azkaban (Book 3)')
```

```python
def recommend(book):
  book = searchBook(book)
  tempdf = books[books['title'].isin(book)]
  return tempdf
```
